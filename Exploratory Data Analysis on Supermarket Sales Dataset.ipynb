# Visualization

### Univariate Analysis

#### 1. Find Aggregate Sales among Branches.

```python
ax = sns.countplot(data = df, x = 'Branch')
ax.bar_label(ax.containers[0], label_type='edge', fontsize = 7, rotation = 45)
plt.show()
```

```python
df['Branch'].value_counts().plot(kind="pie",autopct="%1.2f%%")
plt.show()
```

#### 2. Find the Most Popular Payment Method used by Customers.

```python
df['Payment'].value_counts().plot(kind='barh')
plt.show()
```

#### 3. Find the Distribution of Customer Ratings

```python
sns.distplot(df['Rating'])
plt.show()
```

```python
df['Rating'].skew()
```

#### 4. Find the Distribution of Cost Of Goods Sold.

```python
sns.displot(df['cogs'])
plt.show()
```

```python
df['cogs'].skew()
```

```python
sns.boxenplot(df['cogs'])
plt.show()
```

### Bivariate Analysis / Multivariate Analysis

#### 1. Does the Cost of Goods Sold affect the Ratings that the Customers provide?

```python
sns.scatterplot(data=df, x="cogs", y="Rating")
plt.show()
```

#### 2. Does Gross Income affect the Ratings that the Customers provide?

```python
sns.scatterplot(data=df, x='gross income', y='Rating')
plt.show()
```

#### 3. Find the most Profitable Branch as per Gross Income.

```python
sns.barplot(data=df, x='Branch', y='gross income')
plt.show()
```

```python
sns.barplot(data=df, x='Branch', y='gross income', hue='City')
plt.show()
```

#### 4. Is there any relationship between Gender and Gross Income?

```python
sns.boxplot(data=df, x='Gender', y='gross income')
plt.show()
```

```python
sns.boxplot(data=df, x='Gender', y='gross income',hue='Customer type')
plt.show()
```

#### 5. Find the Product Line that generates the most Income.

```python
df.columns
```

```python
sns.barplot(data=df,x='Product line', y='gross income')
plt.xticks(rotation=80)
plt.show()
```

#### 6. Find the Highest Unit Price in the Product Line.

```python
sns.barplot(data=df,x='Product line', y='Unit price')
plt.xticks(rotation=80)
plt.show()
```

#### 7. Find different Payment methods used by Customers Citywise.

```python
pd.crosstab(df['City'],df['Payment'])
```

```python
sns.heatmap(pd.crosstab(df['City'],df['Payment']))
plt.show()
```

#### 8. Which Product Line is purchased in the Highest Quantity?

```python
(df.groupby('Product line').sum()['Quantity']).plot(kind='bar')
plt.show()
```

#### 9. Find Total Revenue for each Product Line.

```python
df.columns
```

```python
ax = sns.barplot(data = df, x = 'Product line', y = 'Total', errorbar = None, estimator = 'sum')
plt.xticks(rotation = 45)
ax.bar_label(ax.containers[0], label_type='edge', fontsize = 7)
ax.set_title("Total Revenue Each Product Line")
plt.show()
```

#### 10. Find the Count for Male and Female.

```python
ax = sns.countplot(data = df, x = 'Gender')
ax.bar_label(ax.containers[0], label_type='edge', fontsize = 7, rotation = 45)
plt.show()
```
